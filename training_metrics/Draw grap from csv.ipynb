```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('amazon_hmdvr_df_tokenized_sentiment_score_model-ResNet34_epochs-400-batch-512-2023-02-17_10:15:01.csv')
graph_title = 'model-ResNet34_epochs-400-batch-512-2023-02-17_10:15:01'
```

```python
df[['train_mae', 'validation_mae']].plot()
plt.title("Mean Abs Error [STR] : {}".format(graph_title))
plt.ylabel("Mean Abs Error [STR]")
plt.xlabel("Training Epochs")
plt.ylim([0,1.5])
plt.legend()
plt.show()

df[['train_mse', 'validation_mse']].plot()
plt.title("Mean Square Error [$STR^2$] : {}".format(graph_title))
plt.ylabel("Mean Square Error [$STR^2$] ")
plt.xlabel("Training Epochs")
plt.ylim([0,1.5])
plt.legend()
plt.show()


plt.title("Train Accuracy : {}".format(graph_title))
df['train_accuracy'].plot()
#plt.plot(range(1,EPOCHS+1),mse_history["val"],label="validation mse")
plt.ylabel("Train Accuracy")
plt.xlabel("Training Epochs")
#plt.ylim([0,1.5])
plt.legend()
plt.show()
```
