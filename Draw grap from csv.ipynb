```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
log_path = '2023-03-10_151324_jupyter'
hyperparameter_name = 'model-ResNet18_epochs-600-batch-512-randomsplit-True-2023-03-10_151324'
df = pd.read_csv(f'results/{log_path}/training_metrics/amazon_hmdvr_df_tokenized_sentiment_score_{hyperparameter_name}.csv')
graph_title = hyperparameter_name
```

```python
df
```

```python
dff = df.drop(['validation_mae','validation_mse','validation_rmse','validation_mape','validation_mbe','validation_accuracy'], axis=1)
dff.tail()
```

```python
df[['train_mae', 'validation_mae']].plot()
plt.title("Mean Abs Error [STR] : {}".format(graph_title))
plt.ylabel("Mean Abs Error [STR]")
plt.xlabel("Training Epochs")
plt.ylim([0,1.5])
plt.legend()
plt.show()

df[['train_mse', 'validation_mse']].plot()
plt.title("Mean Square Error [$STR^2$] : {}".format(graph_title))
plt.ylabel("Mean Square Error [$STR^2$] ")
plt.xlabel("Training Epochs")
plt.ylim([0,5])
plt.legend()
plt.show()

df[['train_rmse', 'validation_rmse']].plot()
plt.title("Root Mean Squared Error: {}".format(graph_title))
plt.ylabel("Root Mean Squared Error")
plt.xlabel("Training Epochs")
plt.legend()
plt.show()

df[['train_mape', 'validation_mape']].plot()
plt.title("Mean Absolute Percentage Error : {}".format(graph_title))
plt.ylabel("Mean Absolute Percentage Error(%)")
plt.xlabel("Training Epochs")
plt.ylim([0,100])
plt.legend()
plt.show()

df[['train_mbe', 'validation_mbe']].plot()
plt.title("Mean Bias Error : {}".format(graph_title))
plt.ylabel("Mean Bias Error")
plt.xlabel("Training Epochs")
plt.ylim([0,1.5])
plt.legend()
plt.show()

df[['train_accuracy', 'validation_accuracy']].plot()
plt.title("Train Accuracy : {}".format(graph_title))
#plt.plot(range(1,EPOCHS+1),mse_history["val"],label="validation mse")
plt.ylabel("Accuracy")
plt.xlabel("Training Epochs")
#plt.ylim([0,1.5])
plt.legend()
plt.show()
```
